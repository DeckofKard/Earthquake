```python
import folium

# 위도와 경도 데이터
data = [
    {"name": "한빛", "latitude": 35.4089, "longitude": 126.4164},
    {"name": "한울", "latitude": 37.1125, "longitude": 129.3853},
    {"name": "월성", "latitude": 35.7135, "longitude": 129.4772},
    {"name": "새울", "latitude": 35.3357, "longitude": 129.3115},
    {"name": "고리", "latitude": 35.3264, "longitude": 129.2866}
]


# 반경 설정 (단위: 미터)
radius = 30000


# 지도 생성
map = folium.Map(location=[36.5, 128.0], zoom_start=7)  # 초기 위치와 확대 수준 설정

# 마커 추가
for item in data:
    folium.Marker(
        location=[item['latitude'], item['longitude']],
        popup=item['name']
    ).add_to(map)

    
    
for item in data:
    # 마커 추가
    folium.Marker(
        location=[item['latitude'], item['longitude']],
        
        popup=item['name']
    ).add_to(map)

    # 원형 영역 추가
    folium.Circle(
        location=[item['latitude'], item['longitude']],
        radius=radius,
        color='red',
        fill=True,
        fill_opacity=0.3
    ).add_to(map)    

    
    
# 지도를 HTML 파일로 저장
map


```

```python
## 월성
import folium
import pandas as pd
import folium
file_path = "C:/Users\dbdlf/OneDrive/바탕 화면/3-1/빅데이터 프로그래밍/원전 주변 50km 지진 발생 역사/월성 지진 목록.xlsx"
data = pd.read_excel(file_path)

df1 = data.iloc[1:,5:7]
print(df1)

# 시리즈를 리스트로 변화 
series1 = pd.Series(data.iloc[1:,5]).to_list()
series2 = pd.Series(data.iloc[1:,6]).to_list()



# 위도와 경도 데이터
latitudes = series1

longitudes = series2



# 지도 생성
map = folium.Map(location=[36.5, 128.0], zoom_start=10)

# 점 표시
for lat, lon in zip(latitudes, longitudes):
    folium.CircleMarker(location=[lat, lon], radius=5, color='yellow', fill=True, fill_color='').add_to(map)

# 지도 표시
map
```
